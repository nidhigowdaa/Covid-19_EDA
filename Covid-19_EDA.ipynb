```python
tot_cases=df[['State/UTs','Total Cases','Active','Discharged','Deaths']]
```

```python
tot_cases
```

```python
tot_cases_sorted=tot_cases.sort_values(by='Total Cases', ascending=False)
```

```python
top5=tot_cases_sorted.head(5)
top5
```

```python
state_names=top5['State/UTs']
state_values=top5['Total Cases']
```

```python
#top 5 states with the highest number of cases
plt.figure(figsize=(6,6))
plt.pie(state_values,labels=state_names,autopct='%1.2f%%')
```

```python
top5
```

```python
#Total cases and discharged for the 5 states with high number of cases
fig, ax = plt.subplots(figsize=(12, 8))
x = np.arange(len(top5['State/UTs'].unique()))
bar_width = 0.4
b1 = ax.bar(x, top5['Total Cases'],
            width=bar_width)
b2 = ax.bar(x + bar_width, top5['Discharged'],
            width=bar_width)


ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(top5['State/UTs'].unique())

ax.legend(['Total Cases','Discharged'])

for bar in ax.patches:
  
  bar_value = bar.get_height()
  text = f'{bar_value}'
  text_x = bar.get_x() + bar.get_width() / 2 
  text_y = bar.get_y() + bar_value
  bar_color = bar.get_facecolor()
  
  ax.text(text_x, text_y, text, ha='center', va='bottom', color=bar_color,
          size=12)
```

```python
population_df=pd.read_csv('Population.csv') #replacing population values
```

```python
population_df.head()
```

```python
df=df.drop('Population',axis=1)
```

```python
df.head()
```

```python
df1=pd.merge(df,population_df,how='left')
```

```python
df1.head()
```

```python
df1[df1['Population']<df1['Total Cases']]
```

```python
(df1['Total Cases']/df1['Population'])*100
```

```python
df1['case%']=(df1['Total Cases']/df1['Population'])*100
```

```python
df1['active%']=(df1['Active']/df1['Population'])*100
df1['discharged%']=(df1['Discharged']/df1['Population'])*100
df1['death%']=(df1['Deaths']/df1['Population'])*100
```

```python
df1.head()
```

```python
percentages=df1[['State/UTs','case%','active%','discharged%','death%']]
```

```python
percentages.head()
```

```python
#states with high percentage of cases
top_case_perc=percentages.sort_values(by='case%',ascending=False).head(5)
top_case_perc
```

```python
#percentage of total active cases and deaths for the top states with high case percentages
fig, ax = plt.subplots(figsize=(12, 8))
x = np.arange(len(top_case_perc['State/UTs'].unique()))
bar_width = 0.4
b1 = ax.bar(x, top_case_perc['active%'],
            width=bar_width)
b2 = ax.bar(x + bar_width, top_case_perc['death%'],
            width=bar_width)


ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(top_case_perc['State/UTs'].unique())

ax.legend(['active%','death%'])

for bar in ax.patches:
  
  bar_value = bar.get_height()
  text = f'{bar_value}'
  text_x = bar.get_x() + bar.get_width() / 2 
  text_y = bar.get_y() + bar_value
  bar_color = bar.get_facecolor()
  
  ax.text(text_x, text_y, text, ha='center', va='bottom', color=bar_color,
          size=12)
```
